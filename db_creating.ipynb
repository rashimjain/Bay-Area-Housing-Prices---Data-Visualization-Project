```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
```

```python
t=pd.read_csv("CombinedData.csv")
```

```python
t.head()
```

```python
t.fillna(0, inplace=True)
t.head()
```

```python
#start index from 1 instead of 0
t.index += 1
#reset index
t.reset_index(drop=False, inplace=True)
```

```python
#rename 'index' column to 'id'
t.rename(columns={'index':'id'}, inplace=True)
```

```python
#Use declarative_base and create ORM classes for each table.
class DataSet(Base):
    __tablename__='dataset'
    id=Column(Integer, primary_key=True)
    city=Column(String)
    year=Column(String)
    price_per_sqft=Column(Integer)
    monthly_rental_price=Column(Integer)
    forecast_YoY_pct_change=Column(Float)
    latitude=Column(Float)
    longitude=Column(Float)
    population_density=Column(Float)
    violent_crimes=Column(Integer)
    violent_crime_rate=Column(Float)
    property_crime=Column(Integer)
    property_crime_rate=Column(Float)
    population=Column(Integer)
    population_increase=Column(Integer)
    population_increase_pct=Column(Float)
    total_homes=Column(Integer)
    occupied_homes=Column(Integer)
    unoccupied_homes=Column(Integer)
    housing_increase_units=Column(Integer)
    housing_increase_pct=Column(Float)
    single_detached=Column(Integer)
    single_attached=Column(Integer)
    two_to_four_units=Column(Integer)
    five_plus_units=Column(Integer)
    mobile_homes=Column(Integer)
    vacancy_rate_pct=Column(Float)
    persons_per_household=Column(Float)
    ndxt_yr_open_price=Column(Float)
```

```python
#Use the engine and connection string to create a database called hawaii.sqlite
engine = create_engine("sqlite:///project.sqlite")
```

```python
#Once you have your ORM classes defined, create the tables in the database using create_all
Base.metadata.create_all(engine)
```

```python
engine.table_names()
```

```python
#connect to the database
conn=engine.connect()
```

```python
#call the to_sql method to convert df to a table in a database.
#set the 'keep_exists'='append' to  insert data into existing tabels.

t.to_sql("dataset", conn, if_exists="append", index=False)
```

```python
#We can then verify that everything works by querying the database
engine.execute("select * from dataset limit 10").fetchall()
```
